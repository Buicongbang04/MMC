## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id

Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
# ĐIỀN CODE VÀO ĐÂY
result_1 = pd.read_sql("""SELECT m.id, m.season, m.stage, l.name AS league_name, c.name AS country_name
                          FROM match m JOIN League l ON m.league_id = l.id
                                       JOIN Country c ON m.country_id = l.id;""", conn)
result_1
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
# ĐIỀN CODE VÀO ĐÂY
result_2 = pd.read_sql("""SELECT l.name AS league_name, COUNT(m.id) AS total_match
                          FROM match m JOIN League l ON m.league_id = l.id
                          GROUP BY l.name
                          ORDER BY total_match DESC;""", conn)
result_2
```

## Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất
- Sử dung câu lênh GROUP BY, LIMIT

```python
# ĐIỀN CODE VÀO ĐÂY
result_5 = pd.read_sql("""
                      SELECT t.team_long_name, SUM(m.home_team_goal) AS total_home_team_goal
                      FROM Team t LEFT JOIN Match m ON t.team_api_id = m.home_team_api_id
                      GROUP BY t.team_long_name
                      ORDER BY total_home_team_goal DESC
                      LIMIT 20;
                      """, conn)
result_5.head()
```

## Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất
- Sử dụng câu lênh GROUP BY, LIMIT

```python
# ĐIỀN CODE VÀO ĐÂY
result_6 = pd.read_sql("""
                      SELECT t.team_long_name, SUM(m.home_team_goal) AS total_away_team_goal
                      FROM Team t LEFT JOIN Match m ON t.team_api_id = m.away_team_api_id
                      GROUP BY t.team_long_name
                      ORDER BY total_away_team_goal DESC
                      LIMIT 20;
                      """, conn)
result_6.head()
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
# ĐIỀN CODE VÀO ĐÂY
result_7 = pd.read_sql("""
                        SELECT team_long_name, SUM(total_goal) AS total_goal
                        FROM(SELECT t.team_long_name, SUM(m.home_team_goal) AS total_goal
                              FROM Team t LEFT JOIN Match m ON t.team_api_id = m.home_team_api_id
                              GROUP BY t.team_long_name
                              UNION
                              SELECT t.team_long_name, SUM(m.away_team_goal) AS total_goal
                              FROM Team t LEFT JOIN Match m ON t.team_api_id = m.away_team_api_id
                              GROUP BY t.team_long_name) AS t
                        GROUP BY team_long_name    
                        ORDER BY total_goal DESC                   
                       """, conn)
result_7.head()
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
# ĐIỀN CODE VÀO ĐÂY
result_8 = pd.read_sql("""
                        SELECT t.team_long_name, COUNT(m.id) AS total_match
                        FROM Team t JOIN Match m ON t.team_api_id = m.home_team_api_id
                        GROUP BY t.team_long_name
                        ORDER BY total_match DESC
                       """, conn)
result_8.head()
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
# ĐIỀN CODE VÀO ĐÂY
result_9 = pd.read_sql("""
                        SELECT t.team_long_name, SUM(m.home_team_goal) AS total_home_goal, SUM(m.away_team_goal) AS total_away_goal, 
                        CASE  WHEN SUM(m.home_team_goal) > SUM(m.away_team_goal) THEN 'Win'
                              WHEN SUM(m.home_team_goal) = SUM(m.away_team_goal) THEN 'Draw'
                              ELSE 'Lose' 
                        END AS result
                        FROM Team t JOIN Match m ON t.team_api_id = m.home_team_api_id
                        GROUP BY t.team_long_name
                       """, conn)
result_9.head()
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
# ĐIỀN CODE VÀO ĐÂY
result_10 = pd.read_sql("""
                          SELECt team_long_name, COUNT(result) AS total_win
                          FROM (SELECT t.team_long_name, 
                                CASE WHEN m.home_team_goal > m.away_team_goal THEN 'Win'
                                    WHEN m.home_team_goal = m.away_team_goal THEN 'Draw'
                                    ELSE 'Lose' 
                                END AS result
                                FROM Team t JOIN Match m ON t.team_api_id = m.home_team_api_id) AS t
                          WHERE result = 'Win'
                          GROUP BY team_long_name
                          ORDER BY total_win DESC
                          LIMIT 10
                        """, conn)
result_10.head(20)
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
# ĐIỀN CODE VÀO ĐÂY
result_12 = pd.read_sql("""                                       
                        SELECT league_id, total_goal, total_goal * 100 / 70287 AS goal_rate
                        FROM(SELECT league_id, SUM(home_team_goal + away_team_goal) AS total_goal
                             FROM Match
                             GROUP BY league_id) AS t;
                        """, conn)
result_12.head()
```
